```python
# Importing necessary library
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus

# sample text for performing tokenization
text = "In Brazil they drive on the right-hand side of the road. Brazil has a large coastline on the eastern side of South America"

# importing word_tokenize from nltk
from nltk.tokenize import word_tokenize # Passing the string text into word tokenize for breaking the sentences
token = word_tokenize(text)
token
```

```python
text
```

```python
# Finding the frequency distinct in the tokens
# Importing FreqDist library from nltk and passing token into FreqDist
from nltk.probability import FreqDist
fdist = FreqDist(token)
fdist
```

```python
# To find the frequency of top words
fdist1 = fdist.most_common(100)
print('The',len(fdist1),'most common words:')
fdist1
```

```python
import codecs
from collections import Counter
wordDict = Counter()

with codecs.open('El amor en los tiempos del colera.txt','r',encoding='utf8') as f:
    for line in f:
        wordDict.update(line.strip().split())

for word, count in wordDict.most_common(): 
    print(word, count)
```

```python
# importing stopwors from nltk library
from nltk import word_tokenize
from nltk.corpus import stopwords
a = set(stopwords.words('spanish'))
text1 = "Cristiano Ronaldo nació el 5 de Febrero de 1985 en Funchal, Madeira, Portugal."
textl = word_tokenize(text1.lower())
print(textl,len(textl),'\n')
stopwords = [x for x in textl if x not in a]
print(stopwords,len(stopwords))
```
