# Redes: El Gran Dios Pan

#### Se procederá a analizar la obra de Arthur Machen - "El Gran Dios Pan" y se creará una red de relaciones entre los personajes, posteriormente se determinara si es de mundo pequeño, libre escala  o ambas.

#### Para mayor facilidad se analizará capitulo a capitulo, presentando al final la red final



```python
#Función obtención de frecuencia y relaciones entre personajes 15 palabras despues de la primera aparición

def FR(personaje1,personaje2,archivo):
    m = len(archivo)
    
    #print(m)
    
    FP = 0
    RP = 0
    
    for i in range(m):
        if archivo[i] == personaje1:
            #print('coincidencia')
            FP = FP + 1
        
            for w in range(14):
                k = w +1
            
                if i+k == m-1:
                    break
            
                if archivo[i+k] == personaje2:
                    #print('relación')
                    RP = RP + 1
                
    return FP, RP
```

## Extraer la obra de PDF a TXT

```python
#Librería para pasar de un archivo PDF a un archivo TXT
import PyPDF2
```

```python
#Función para extraer el contenido del PDF

def getPDFtoTXT(pdffile):
    myPDFFile = PyPDF2.PdfFileReader(pdffile)
    
    with open('./pdfContentPan.txt','w',encoding='utf-8', errors='ignore') as pdf_output:
        for page in range(myPDFFile.getNumPages()):
            data = myPDFFile.getPage(page).extractText()
            pdf_output.write(data)
            
    with open('./pdfContentPan.txt','r',encoding='utf-8', errors='ignore') as myPDFContent:
        return myPDFContent.read().replace('\n',' ')
```

## Análisis de la Obra 

#### Se analizará las fecuencia de aparación de los personajes y se guardará en una variable entera la relación entre los personajes con la siguiente regla de relación:  Existe una relación entre personajes, si después de aparecer un personaje A aparece un personaje B dentro de las siguientes 15 palabras en el texto. 

```python
#Librería para analizar la obra en TXT
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

#Primero abrimos el archivo correspondiente al primer capítulo
file = open("C1Pan.txt", "r",encoding='utf-8')
C1_original = file.read()

#Para evitar ambigüedades del texto original, todo lo pondremos en minúsculas 
tw1= word_tokenize(C1_original.lower())
#print(tw1)

#Quitaremos los puntos, comas, signos de admiración, exclamación, guiones que no corresponden a una palabra como tal
text2=""". ¿ , ? y ; " " : ¿y ¡ ! """


stop_words=word_tokenize(text2)
#print(stop_words)

C1_filtrado=[]
for w in tw1:
    if w not in stop_words:
        C1_filtrado.append(w)
#print("Tokenized Sentence:",tw1)
print("Filterd Sentence:",C1_filtrado)

```

```python
import numpy as np

#Lista de Personajes 
C1_Personajes = ['clarke','raymond','mary','dios','helen','trevor','rachel','herbert','villiers','austin','meyrick','argentine','collier','herries','crashaw']
C1_PersonajesT = ['clarke','raymond','mary','dios','helen','trevor','rachel','herbert','villiers','austin','meyrick','argentine','collier','herries','crashaw']
d = len(C1_Personajes)

ApodosC1 = np.zeros(d)
NodosC1 = np.zeros(d)

C1_Matriz = np.zeros((d,d))
C1_MatrizSUM = np.zeros((d,d))

#Ciclo para obtener
for i in range(d):
    for j in range(d):
        
        u,v = FR(C1_Personajes[i],C1_PersonajesT[j],C1_filtrado)
        C1_Matriz[i,j] = v
        
        if j == 1:
            ApodosC1[i] = u


for i in range(d):
    for j in range(d):
        if i == j:
            C1_MatrizSUM[i,j] = C1_Matriz[i,j]
        elif i<j:
            C1_MatrizSUM[i,j] = C1_Matriz[i,j] + C1_Matriz[j,i]
        else: 
            C1_MatrizSUM[i,j] = C1_MatrizSUM[j,i]
            
            
print("Matriz de Adyecencias de todos los personajes\n")
print(C1_MatrizSUM)

print("\nFrecuencia de Aparición de Personajes\n")
print(ApodosC1)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

A = ApodosC1*10
node_sizes = A
G = nx.Graph()

for i in range(d):
    for j in range(d):
        if i == j:
            G.add_edge(C1_Personajes[i], C1_PersonajesT[j], weight=0)
        else:
            G.add_edge(C1_Personajes[i], C1_PersonajesT[j], weight=C1_Matriz[i,j])
        
        #G.add_edge(C1_Personajes[i], C1_PersonajesT[j], weight=(C1_MatrizSUM[i,j]/14)*5)
        
ws = [d['weight'] for u,v,d in G.edges(data=True)]

#plt.figure(figsize=(10,10),dpi=150)
plt.figure(figsize=(15,15))
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)

pos =nx.spring_layout(G)
#nx.rescale_layout_dict(pos, scale=10)

# nodes
nx.draw_networkx_nodes(G, pos, node_size=A)

# edges
nx.draw_networkx_edges(G, pos, width=ws)

# labels
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")


plt.axis("off")
plt.show()

nx.eccentricity(G)
```

Funciones para estimar la pequeñez de los gráficos.

Una red mundial pequeña se caracteriza por una longitud de camino más corta promedio pequeña y un coeficiente de agrupamiento grande.

La pequeñez se mide comúnmente con el coeficiente sigma u omega.

#### Vamos a utilizar omega

omega = Lr / L - C / Cl

donde C y L son respectivamente el coeficiente de agrupamiento promedio y la longitud de camino más corta promedio de G. Lr es la longitud de camino más corta promedio de un gráfico aleatorio equivalente y Cl es el coeficiente de agrupamiento promedio de un gráfico reticular equivalente.

El coeficiente de mundo pequeño (omega) varía entre -1 y 1. Los valores cercanos a 0 significan que G presenta características de mundo pequeño. Los valores cercanos a -1 significan que G tiene una forma de celosía, mientras que los valores cercanos a 1 significan que G es un gráfico aleatorio.

```python
nx.omega(G, niter=100, nrand=10, seed=None)
```

#### Notamos que el valor de omega es identicamente cero, por lo que concluimos que nuestra red es de mundo pequeño que se reafirma por la red anterioremente graficada
