```python
#Función obtener vector de bites
#Esta funcion recibe un vector de la matriz de la imagen en escala de grises 

def vector(renglon):
    by = list(map(int,format(renglon[0], "#010b")[2:]))

    for i in range(tx-1):
        by2 = list(map(int,format(renglon[i+1], "#010b")[2:]))
        by = by + by2


    vect1 = np.array(by)
    
    return vect1
```

```python
#Función Iteración Encriptación
#Esta función recibe dos vectores y los itera 10 veces 


def iteracion(vect1,vect2):
    
    N = 10
    txb = tx*8
    s = np.zeros(txb)
    A = np.zeros((N+2,txb))
    
    A[0,:] = vect1
    A[1,:] = vect2
    k=1
    
    for w in range(N):
        
        k = k + 1
        
        #print(vect1)
        #print(vect2)
        
        for i in range(txb):
    
            j = i
    
            if(j==0):
                s[j] = ((1)*vect2[txb-1] + (0)*vect2[j] + (1)*vect2[j+1] + vect1[j])%2
            elif(i==(txb-1)):
                s[j] = ((1)*vect2[j-1] + (0)*vect2[j] + (1)*vect2[0] + vect1[j])%2 
            else:
                s[j] = ((1)*vect2[j-1] + (0)*vect2[j] + (1)*vect2[j+1] + vect1[j])%2 
              
        #print(s)
        A[k,:]= s
        
        vect1 = vect2
        vect2 = s
      

    return A
    
```

```python
#Funcion division de vectores de bites en 8 para transformacion a binario
#Me regresa un gris renflon de la matriz cifrada

def div(vector):
    
    GrisB = np.zeros((tx,8))
    GrisR = np.zeros(tx)
    
    byt = np.zeros(8)
    ww = 0
    
    for i in range(tx):
        
        for j in range(8):
            
            GrisB[i,j] = vector[ww]
            
            #print(vector[ww])
            ww = ww + 1
            
    for i in range(tx):
        for j in range(8):
        
            GrisR[i] = GrisR[i] + GrisB[i,j]*(2**(7-j))
            
            
    return GrisR
```

```python
#Función que crea una matriz de grises con los valores enteros

def gris(A):
    
    B = np.zeros((ta,tx,3))
    
    for i in range(ta):
        for j in range(tx):
            
            B[i,j,0]=A[i,j]
            B[i,j,1]=A[i,j]
            B[i,j,2]=A[i,j]
    
    return B

```

### Aqui empieza la encriptación de la matriz imagen

```python
#Primero vamos a convertir la matriz de grises en una matriz de bytes

Mby = np.zeros((ta,tx*8))

for i in range(ta):
    v = MatINO[i,:]
    Mby[i,:] = vector(v)
    
    
print(Mby)
```

```python
#Vamos a iterar los renglones y a construir la matriz cifrada de bytes

taI = int (ta/2)

MbyC = np.zeros((ta,tx*8))
k=0

for i in range(taI):
    vv1 = Mby[k,:]
    vv2 = Mby[k+1,:]

    E= iteracion(vv1,vv2)
    
    MbyC[k,:] = E[10,:]
    MbyC[k+1,:] = E[11,:]
    
    k = k + 2

print(MbyC)
```

```python
#Vamos a recuperar los numeros enteros de la matriz 

MC = np.zeros((ta,tx))


for i in range(ta):
    m = div(MbyC[i,:])
    
    MC[i,:] = m
    
    
for i in range(ta):
    for j in range(tx):
        
        MC[i,j] = int(MC[i,j])
    
#Crea la matriz de grises cifrado   
#Creo la función 

MCF = gris(MC)
MCC = MCF.astype(int)

print(MCC) 
```
