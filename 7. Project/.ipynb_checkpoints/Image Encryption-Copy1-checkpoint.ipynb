```python
import matplotlib.pyplot as plt
import numpy as np
import sys, pygame
pygame.init()

#Tama√±o de pantalla
size = width, height = 500, 500

#N√∫mero de celdas
n = 100
nX = n
nY = n

dimCW = (width - 1)/nX
dimCH = (height- 1)/nY

bg = 25, 25, 25

screen = pygame.display.set_mode(size)
screen.fill(bg)

#gameState = np.random.randint(0, 2, (nX, nY))

gameState = np.zeros((nX, nY))
gameState[int(n/2)-1,int(n/2)-1] = 1
gameState[int(n/2)-1,int(n/2)  ] = 1
gameState[int(n/2)-1,int(n/2)+1] = 1
gameState[int(n/2)  ,int(n/2)-1] = 1
gameState[int(n/2)  ,int(n/2)  ] = 1
gameState[int(n/2)  ,int(n/2)+1] = 1
gameState[int(n/2)+1,int(n/2)-1] = 1
gameState[int(n/2)+1,int(n/2)  ] = 1
gameState[int(n/2)+1,int(n/2)+1] = 1

#print(gameState)

#Ciclo para crear las celdas y la malla

#Fronteras peri√≥dicas

for i in range(26): #62 o 126 iteraciones est√°n buenas
    
    pygame.event.pump()
    
    new_gameState = np.copy(gameState)
    screen.fill(bg)
    
    for y in range(0, nY):
        for x in range(0, nX):
            
            #Regla de Solidificaci√≥n:
            #1. Si la celda est√° activa (estado 1) permanece activa
            #2. Si la celda est√° inactiva (estado 0) entonces se suman los estados de las celdas vecinas,
            #   si el n√∫mero de celdas activadas en su entorno es 1 o 2, entonces la celda se activa,
            #   en caso contrario permanecer√° desactivada
           
            ######FUNCI√ìN PARA VER CUANTOS VECINOS DE [X,Y] ESTAN VIVOS
            
#             n_neigh = gameState[(x-1) % nX, (y-1) % nY ] + \
#                       gameState[(x)   % nX, (y-1) % nY ] + \
#                       gameState[(x+1) % nX, (y-1) % nY ] + \
#                       gameState[(x-1) % nX, (y)   % nY ] + \
#                       gameState[(x)   % nX, (y)   % nY ] + \
#                       gameState[(x+1) % nX, (y)   % nY ] + \
#                       gameState[(x-1) % nX, (y+1) % nY ] + \
#                       gameState[(x)   % nX, (y+1) % nY ] + \
#                       gameState[(x+1) % nX, (y+1) % nY ]
            
            n_neigh = 1*(gameState[(x-1) % nX, (y-1) % nY ] + \
                      gameState[(x+1) % nX, (y-1) % nY ] + \
                      gameState[(x-1) % nX, (y+1) % nY ] + \
                      gameState[(x+1) % nX, (y+1) % nY ]) + \
                      gameState[(x)   % nX, (y-1) % nY ] + \
                      gameState[(x-1) % nX, (y)   % nY ] + \
                      gameState[(x)   % nX, (y)   % nY ] + \
                      gameState[(x+1) % nX, (y)   % nY ] + \
                      gameState[(x)   % nX, (y+1) % nY ]
            
            ######APLICACI√ìN DE LAS REGLAS
            
            if gameState[x,y] == 1:
                new_gameState[x,y] = 1
                
            elif gameState[x,y] == 0 and (n_neigh == 1 or n_neigh == 2):
                new_gameState[x,y] = 1
            
            poly = [
                    ((x)*dimCW,   (y)*dimCH),
                    ((x+1)*dimCW, (y)*dimCH),
                    ((x+1)*dimCW, (y+1)*dimCH),
                    ((x)*dimCW,   (y+1)*dimCH)
                    ]
            
            pygame.draw.polygon(screen, (128, 128, 128), poly, int(abs(1-new_gameState[x,y])))

    gameState = new_gameState
        
    plt.rcParams['image.cmap'] = 'binary'
    plt.matshow(gameState)
    plt.axis('off')
    plt.show()         
    
#     pygame.display.flip()
```

```python
def circular(T, Œª_1=1, Œª0=0, Œª1=1): #T es el n√∫mero de iteraciones
    
    n = 8*data.shape[1]-1
    J = np.zeros(data.shape[1],int) #Para guardar en una matriz de enteros; al final borro este rengl√≥n
    
    for k in range(0,data.shape[0],2): #Renglones

        cT_2 = []
        cT_1 = []
        for i in range(data.shape[1]): #Columnas

            string = np.binary_repr(data[k,i], 8)
            cT_2 += [int(bit) for bit in string]

            string = np.binary_repr(data[k+1,i], 8)
            cT_1 += [int(bit) for bit in string]

        cT = []

        for j in range(2,T+1):
            for i in range(n):
                cT.append((cT_2[i] + Œª_1*cT_1[i-1] + Œª0*cT_1[i] + Œª1*cT_1[i+1])%2) #Regla de actualizaci√≥n
            cT.append((cT_2[n] + Œª_1*cT_1[n-1] + Œª0*cT_1[n] + Œª1*cT_1[0])%2) #√öltimo elemento con condiciones peri√≥dicas
            cT_2 = cT_1.copy()
            cT_1 = cT.copy()
            cT = []

        rT_2 = []
        rT_1 = []
        for j in range(0,len(cT_2),8):
            num0 = 0
            num1 = 0
            for i in range(8):
                num0 += cT_2[i+j]*2**(7-i)
                num1 += cT_1[i+j]*2**(7-i)
            rT_2.append(num0)
            rT_1.append(num1)
            
        J = np.vstack([J,[rT_2],[rT_1]])
        
    J = np.delete(J,0,0)

#     img.imsave('J.jpg',J)
    plt.imshow(J,cmap='cividis')
    plt.title('T={}\n({},{},{})'.format(T, Œª_1, Œª0, Œª1))
    plt.axis('off')
    
#     return J
```

```python
def cifrado(T, Œª_1=1, Œª0=0, Œª1=1): #T es el n√∫mero de iteraciones
#     J = [] #Para guardar en una lista
    n = 8*I.shape[1]-1
    J = np.zeros(I.shape[1],int) #Para guardar en una matriz de enteros; al final borro este rengl√≥n
    
    for k in range(0,I.shape[0],2): #Renglones

        cT_2 = []
        cT_1 = []
        for i in range(I.shape[1]): #Columnas

            string = np.binary_repr(I[k,i], 8)
            cT_2 += [int(bit) for bit in string]

            string = np.binary_repr(I[k+1,i], 8)
            cT_1 += [int(bit) for bit in string]

        cT = []

        for j in range(2,T+1):
            for i in range(n):
                cT.append((cT_2[i] + Œª_1*cT_1[i-1] + Œª0*cT_1[i] + Œª1*cT_1[i+1])%2) #Regla de actualizaci√≥n
            cT.append((cT_2[n] + Œª_1*cT_1[n-1] + Œª0*cT_1[n] + Œª1*cT_1[0])%2) #√öltimo elemento con condiciones peri√≥dicas
            cT_2 = cT_1.copy()
            cT_1 = cT.copy()
            cT = []

        rT_2 = []
        rT_1 = []
        for j in range(0,len(cT_2),8):
            num0 = 0
            num1 = 0
            for i in range(8):
                num0 += cT_2[i+j]*2**(7-i)
                num1 += cT_1[i+j]*2**(7-i)
            rT_2.append(num0)
            rT_1.append(num1)

# Para guardar como lista:
#         J.append(rT_2)
#         J.append(rT_1)
            
# Otra forma de guardar como matriz:
#         np.append(J,rT_2,axis=0)
#         np.append(J,rT_1,axis=0)
        
        J = np.vstack([J,[rT_2],[rT_1]])
        
    J = np.delete(J,0,0)
    
#     print(J[0][:3])
#     print(J[1][:3])

    img.imsave('J.jpg',J)
    plt.imshow(J,cmap='gray')
    plt.title('T={}\n({},{},{})'.format(T, Œª_1, Œª0, Œª1))
    plt.axis('off')
    
    return J
```

```python
def descifrado(T, Œª_1=1, Œª0=0, Œª1=1): #T es el n√∫mero de iteraciones
    I = []
    n = 8*J.shape[1]-1
    
    for k in range(0,J.shape[0],2): #renglones hacia adelante

        CT = []
        CT_1 = []

        for i in range(J.shape[1]): #columnas hacia adelante

            string = np.binary_repr(J[k+1,i], 8)   #Rengl√≥n n
            CT += [int(bit) for bit in string]

            string = np.binary_repr(J[k,i], 8) #Rengl√≥n n-1
            CT_1 += [int(bit) for bit in string]            
            
        CT_2 = []

        for j in range(2,T+1):
            for i in range(n):
                CT_2.append((CT[i] + Œª_1*CT_1[i-1] + Œª0*CT_1[i] + Œª1*CT_1[i+1])%2) #Regla de actualizaci√≥n
            CT_2.append((CT[n] + Œª_1*CT_1[n-1] + Œª0*CT_1[n] + Œª1*CT_1[0])%2) #√öltimo elemento con condiciones peri√≥dicas

            CT = CT_1.copy()
            CT_1 = CT_2.copy()
            CT_2 = []

        RT = []
        RT_1 = []
        for j in range(0,len(CT),8):
            numn = 0
            numn_1 = 0
            for i in range(8):
                numn += CT[i+j]*2**(7-i)
                numn_1 += CT_1[i+j]*2**(7-i)
            RT.append(numn)
            RT_1.append(numn_1)
       
        I.append(RT_1)
        I.append(RT)
    
#     print(I[0][:3])
#     print(I[1][:3])

    plt.imshow(I,cmap='gray')
    plt.title('T={}\n({},{},{})'.format(T, Œª_1, Œª0, Œª1))
    plt.axis('off')
```
